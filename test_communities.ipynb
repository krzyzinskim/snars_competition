```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
graphs = os.listdir("competition")
for graph in graphs:

    adjecency_matrix = np.loadtxt(
        f"competition/{graph}", delimiter=","
    )
    G = nx.from_numpy_array(adjecency_matrix)
    
    communities = pd.read_csv(f"results/good/{graph}", delimiter=",", header = None)    

    pos = nx.spring_layout(G, k = 10, iterations = 5000, seed = 42069)
    # pos = nx.kamada_kawai_layout(G)
    nx.draw_networkx_nodes(G, pos, node_size=50, node_color=communities[1], cmap='tab20')
    nx.draw_networkx_edges(G, pos, alpha=0.3)

    if (kstring := graph.split(".")[0].split("-")[1]) == "UNC":
            k = "UNC"
    else:
            k = int(kstring.split("=")[1])

    plt.title(f"k={k}")
    plt.show()

```
